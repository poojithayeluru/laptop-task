```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df1=pd.read_csv("l.csv")
```

```python
df1.head()
```

```python
df1=df1.drop(["Unnamed: 0"],axis=1)
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1["Warranty"]=df1["Warranty"].astype("object")

df1["RAM"]=df1["RAM"].astype("object")
```

```python
xx=df1.drop(["MRP"],axis=1)
yy=df1["MRP"].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y1_train, y1_test = train_test_split(xx, yy, test_size=0.2,random_state=42)
```

```python
X_train.isna().sum()
```

```python
## Applying label encoding for all categorical columns
```

```python
ramencoder={32:4,16:3,8:2,4:1}

X_train["RAM"]=X_train["RAM"].apply(lambda x : ramencoder[x])
```

```python
processorencoder={"M1":14,"i9":13,"M2":12,"Ryzen 9":11,"i7":10,"Ryzen 5":9,"Ryzen 7":8,"i5":7,"i3":6,"Ryzen 3":5,"Athlon Dual":4,"Celeron Dual":3,"Celeron Quad":2,"Pentium":1}

X_train["processor_type"]=X_train["processor_type"].apply(lambda x : processorencoder[x])
```

```python
driveencoder={"Both":4,"SSD":3,"EMMC":2,"HDD":1}

X_train["disk_drive"]=X_train["disk_drive"].apply(lambda x : driveencoder[x])
```

```python
diskencoder={"2":11,"1,512":10,"1,256":9,"256,256":8,"1 SSD":7,"512":6,"1 HDD":5,"256":4,"128":3,"64":2,"32":1}
X_train["Disc_size"]=X_train["Disc_size"].apply(lambda x : diskencoder[x])
```

```python
priceencoder={"low":1,"medium":2,"High":3}
X_train["Price_range"]=X_train["Price_range"].apply(lambda x : priceencoder[x])
```

```python
warantyencoder={1:1,2:2,3:3}
X_train["Warranty"]=X_train["Warranty"].apply(lambda x : warantyencoder[x])
```

```python
## Applying label encoding for test data
```

```python
X_test["RAM"]=X_test["RAM"].apply(lambda x : ramencoder[x])

X_test["processor_type"]=X_test["processor_type"].apply(lambda x : processorencoder[x])

X_test["disk_drive"]=X_test["disk_drive"].apply(lambda x : driveencoder[x])

X_test["Disc_size"]=X_test["Disc_size"].apply(lambda x : diskencoder[x])

X_test["Price_range"]=X_test["Price_range"].apply(lambda x : priceencoder[x])

X_test["Warranty"]=X_test["Warranty"].apply(lambda x : warantyencoder[x])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
```

```python
#Random forest
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',StandardScaler(),[3])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=220,
                              random_state=0,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=12)

p = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y1_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y1_test,y_pred))
print('MAE',mean_absolute_error(y1_test,y_pred))
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',StandardScaler(),[3])
],remainder='passthrough')

step2 = Lasso(alpha=0.001)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y1_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y1_test,y_pred))
print('MAE',mean_absolute_error(y1_test,y_pred))
```
